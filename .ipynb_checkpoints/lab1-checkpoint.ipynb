## Question 1

### a.

```python
import pandas as pd
import numpy as np
df=pd.read_csv('question1.csv', index_col=0)
df.head()
```

```python
first_half=df.iloc[:200]
second_half=df.iloc[200:]
```

### b.

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
x1=first_half['customers'].values.reshape(-1, 1)
y1=first_half['sales']
x2=second_half['customers'].values.reshape(-1, 1)
y2=second_half['sales']
lr.fit(x1,y1)
```

### c.


```python
from scipy.stats import poisson, chi2
customer_1=first_half['customers']
customer_2=second_half['customers']
mu1=customer_1.mean()
mu2=customer_2.mean()
dist1=poisson(mu1)
dist2=poisson(mu2)
```

```python
likelihood1=np.sum(np.log(dist1.pmf(customer_1)))
likelihood2=np.sum(np.log(dist2.pmf(customer_2)))
log_RL=likelihood1-likelihood2
t_stat=-2*log_RL
p_value=1-chi2(df=1).cdf(t_stat)
print(p_value)
```

since p value is very small, the null hypothesis rejected, data drifted.

### d.

```python
from sklearn.metrics import mean_squared_error
pred=lr.predict(x2)
mse=mean_squared_error(y2, pred)
print(mse)
```

Judging by the MSE, the fitted model performed very bad on the sceond half of the data 

## Question 2


### 1. 

```python
df=pd.read_csv("question2.csv")
wool_dummies=pd.get_dummies(df['wool'], dtype=int)
tension_dummies=pd.get_dummies(df['tension'], dtype=int)
```

```python
x=pd.concat([wool_dummies, tension_dummies, df['temperature']], axis=1)
y=df['breaks']
x.head()
```

### 2.

```python
import statsmodels.api as sm
```

```python
glm_model=sm.GLM(y, x, family=sm.families.Poisson())
results=glm_model.fit()
results.summary()
```

Judging by the p values of coefficient estimates, the causes of yarn breakage exclude temperature. 

### 3.


```python
glm_wool=sm.GLM(y, wool_dummies, family=sm.families.Poisson())
glm_tension=sm.GLM(y, tension_dummies, family=sm.families.Poisson())
results_wool=glm_wool.fit()
results_tension=glm_tension.fit()
bic_comparison=pd.Series({'Full model BIC': results.bic, 'Wool BIC': results_wool.bic, 'Tension BIC': results_tension.bic})
bic_comparison
```

Compared to the full model, separate breakage regressions against tesion and wool do not fit well.

### 1.

```python
import altair as alt
df=pd.read_csv("question3.csv")
df.head()
```

```python
df['duration(h)']=(pd.to_datetime(df['pt_out_room_time'])-pd.to_datetime(df['pt_in_room_time'])).dt.seconds/3600
#duration_df=df.groupby(['actual_operating_location'])['duration(h)'].mean() return pandas series
```

```python
charts=[]
for col in df.columns[: 5]:
    duration_df=df.groupby(col).aggregate({'duration(h)': 'mean'}).reset_index()
    new_col=col.replace('_', ' ').title()
    chart=alt.Chart(duration_df, title=f'Mean Duration(h) by {new_col}').mark_bar().encode(
        x=alt.X(col, title=new_col),
        y=alt.Y(duration_df.columns[1], title=duration_df.columns[1].capitalize())
    )
    charts.append(chart)

final_chart=alt.hconcat(*charts)
final_chart
```

```python
charts=[]
duration_df=df.groupby(df.iloc[:, 3].str.extract(r'(\D+)', expand=False)).aggregate({'duration(h)': 'mean'}).reset_index()
chart=alt.Chart(duration_df, title=f'Mean Duration(h)').mark_bar().encode(
        x=alt.X(duration_df.columns[0]),
        y=alt.Y(duration_df.columns[1], title=duration_df.columns[1].capitalize())
    )
charts.append(chart)

duration_df=df.groupby(df.iloc[:, 4].apply(pd.to_datetime).dt.month_name()).aggregate({'duration(h)': 'mean'}).reset_index()
chart=alt.Chart(duration_df, title=f'Mean Duration(h)').mark_bar().encode(
        x=alt.X(duration_df.columns[0]),
        y=alt.Y(duration_df.columns[1], title=duration_df.columns[1].capitalize())
    )
charts.append(chart)

alt.hconcat(*charts)
```

I see 'surgeon_identifier', 'procedure_identifier' have significantly different average durations so I expect them to be meaningful predictors. 

### 2.

```python
hist=alt.Chart(df).mark_bar().encode(
    x=alt.X('duration(h)', bin=alt.Bin(maxbins=20)),
    y='count()'
)
hist
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
data=df['duration(h)']
norm_params=stats.norm.fit(data)
fitted_norm_dist=stats.norm(*norm_params)
plt.hist(data, bins=30, density=True, label='Histogram')
x=np.linspace(min(data), max(data), 100)
plt.plot(x, fitted_norm_dist.pdf(x), label='Fitted Normal Distribution')
plt.xlabel('Value')
plt.ylabel('density')
plt.legend()
plt.show()
```

```python
inv_norm_params=stats.invgauss.fit(data)
fitted_invnorm_dist=stats.invgauss(*inv_norm_params)
plt.hist(data, bins=30, density=True, label='Histogram')
x=np.linspace(min(data), max(data), 100)
plt.plot(x, fitted_invnorm_dist.pdf(x), label='Fitted Inverse Gaussian Distribution')
plt.xlabel('Value')
plt.ylabel('density')
plt.legend()
plt.show()
```

```python
gamma_params=stats.gamma.fit(data)
fitted_gamma_dist=stats.gamma(*gamma_params)
plt.hist(data, bins=30, density=True, label='Histogram')
x=np.linspace(min(data), max(data), 100)
plt.plot(x, fitted_gamma_dist.pdf(x), label='Fitted Gamma Distribution')
plt.xlabel('Value')
plt.ylabel('density')
plt.legend()
plt.show()
```

Judging by the density plot and fitted distribution plot, inverse gaussian distribution is the most appropriate. 

### 3. 

```python
x1_dummies=pd.get_dummies(df.iloc[:, 2], dtype=int)
x2_dummies=pd.get_dummies(df.iloc[:, 3].str.extract(r'(\D+)', expand=False), dtype=int)
#x3_dummies=pd.get_dummies(df.iloc[:, 4].apply(pd.to_datetime).dt.day_name())
x=pd.concat([x1_dummies, x2_dummies], axis=1)
y=df['duration(h)']# inverse gaussian
x.head()
```

```python
inv_gauss_glm = sm.GLM(y, x, family=sm.families.InverseGaussian(link=sm.families.links.Log()))
results=inv_gauss_glm.fit()
predictions=results.predict()
```

```python
mse=mean_squared_error(y, predictions)
mse
```

```python
plt.scatter(y, predictions, label="GLM Predictions")
plt.plot(y, y, color='red', label='Perfect Fit')
plt.xlabel('Actual y')
plt.ylabel('Predicted y')
plt.legend()
plt.show()
```
